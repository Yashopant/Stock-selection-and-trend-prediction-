```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

def beta(ticker_symbol, start_date, end_date, i, window_size=30):
    data = yf.download(ticker_symbol, start=start_date, end=end_date)
    data_nifty = yf.download('^NSEI', start=start_date, end=end_date)
    
    beta_values = []
    
    for j in range(i, i + window_size):
        data_slice = data.iloc[j:j+window_size]
        data_nifty_slice = data_nifty.iloc[j:j+window_size]

        result_df = pd.merge(left=data_slice, right=data_nifty_slice, on='Date', how='outer')
        finaldf = result_df.dropna()
        data_needed = finaldf[['Adj Close_x', 'Adj Close_y']]

        def calculate_percentage_change(data, column_name):
            data['Percentage Change ' + column_name] = data[column_name].pct_change()

        calculate_percentage_change(data_needed, 'Adj Close_x')
        calculate_percentage_change(data_needed, 'Adj Close_y')

        correlation = data_needed['Percentage Change Adj Close_x'].corr(data_needed['Percentage Change Adj Close_y'])
        std_x = np.std(data_needed['Percentage Change Adj Close_x'])
        std_y = np.std(data_needed['Percentage Change Adj Close_y'])

        beta_value = correlation * (std_x / std_y)
        beta_values.append(beta_value)
        
    plt.figure(figsize=(10, 6))
    plt.plot(range(len(beta_values)), beta_values, label=ticker_symbol, marker='o', linestyle='-', color='b')
    plt.grid(True)
    plt.show()
        
    return beta_values
```

```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

def beta_values(ticker_symbol, start_date, end_date, i, j):
    def beta(i):
        data = yf.download(ticker_symbol, start=start_date, end=end_date)
        data_nifty = yf.download('^NSEI', start=start_date, end=end_date)

        data = data.iloc[i:i+30]
        data_nifty = data_nifty.iloc[i:i+j]
        result_df = pd.merge(left=data, right=data_nifty, on='Date', how='outer')
        finaldf = result_df.dropna()
        data_needed = finaldf[['Adj Close_x', 'Adj Close_y']]

        def calculate_percentage_change(data, column_name):
            data['Percentage Change ' + column_name] = data[column_name].pct_change()

        calculate_percentage_change(data_needed, 'Adj Close_x')
        calculate_percentage_change(data_needed, 'Adj Close_y')

        correlation = data_needed['Percentage Change Adj Close_x'].corr(data_needed['Percentage Change Adj Close_y'])
        std_x = np.std(data_needed['Percentage Change Adj Close_x'])
        std_y = np.std(data_needed['Percentage Change Adj Close_y'])

        # Check if std_y is not zero before division
        if std_y != 0:
            beta_value = correlation * (std_x / std_y)
            return beta_value
        else:
            return None  # You can handle this case according to your needs

    beta_values = []

    for i in range(len(data_needed['Percentage Change Adj Close_x'] - 30)):
        beta_val = beta(i)
        if beta_val is not None:
            beta_values.append(beta_val)

    plt.figure(figsize=(10, 6))
    plt.plot(range(len(beta_values)), beta_values, label=ticker_symbol, marker='o', linestyle='-', color='b')
    plt.grid(True)
    plt.show()

    return beta_values


```
